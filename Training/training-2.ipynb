```python
import yaml, shutil

# Just copy the yaml file only — not the whole dataset
shutil.copy(
    '/kaggle/input/datasets/vishweshvishwakarma/geo-dataset/yolo_split/dataset.yaml',
    '/kaggle/working/dataset.yaml'
)

# Fix the yaml to point back to input folder
yaml_path = '/kaggle/working/dataset.yaml'
with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

data['path']  = '/kaggle/input/datasets/vishweshvishwakarma/geo-dataset/yolo_split/'
data['train'] = 'train/images'
data['val']   = 'val/images'
data['test']  = 'test/images'

with open(yaml_path, 'w') as f:
    yaml.dump(data, f)
print("Done!")
print(open(yaml_path).read())
```

```python
from ultralytics import YOLO

model = YOLO('/kaggle/input/models/vishweshvishwakarma/yolo26m-custo/pytorch/default/1/best.pt')  # medium model for better attention/context

results = model.train(
    data='/kaggle/working/dataset.yaml',
    task='segment',
    cache=False,
    epochs=10,
    imgsz=768,       
    batch=16,           # smaller batch for larger tiles on T4 x2
    device=[0, 1],
    optimizer='AdamW',
    lr0=0.0001,         # fresh start with medium model
    lrf=0.01,
    cos_lr=True,
    warmup_epochs=0,
    mosaic=0.0,
    fliplr=0.5,
    flipud=0.5,
    patience=15,
    save=True,
    save_period=3,
    project='/kaggle/working/yolo26_m_seg_project',
    name='run_v3',
    exist_ok=True,
)
```

```python
import shutil
shutil.copy(
    '/kaggle/working/yolo26_m_seg_project/run_v1/weights/best.pt',
    '/kaggle/working/best_v3.pt'
)
print("Saved!")

import shutil
shutil.copy(
    '/kaggle/working/yolo26_m_seg_project/run_v1/weights/last.pt',
    '/kaggle/working/last_v3.pt'
)
print("Saved!")
```
